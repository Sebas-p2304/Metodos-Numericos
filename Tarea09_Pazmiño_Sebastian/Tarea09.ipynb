**5. Dado el sistema lineal:**<br>
<center>

$x_1 - x_2 + \alpha x_3 = -2\\$
$-x_1 + 2x_2 -\alpha x_3 = 3\\$
$\alpha x_1 + x_2 + x_3 = 2$

</center>

**a. Encuentra el valor(es) de $\alpha$ para los que el sistema no tiene solución**
**b. Encuentre los valor(es) de $\alpha$ para los que el sistema tiene un número infinito de soluciones.**
Del sistema de ecuaciones tenemos la siguiente matriz A:

$\left(\begin{array}{cc}
1 & -1 & \alpha\\
-1 & 2 & -\alpha\\
\alpha & 1 & 1
\end{array} \right)$

Si $det(A) = 0$, entonces el sistema tiene infinitas soluciones o ninguna.
Operando:

\begin{equation*}
\begin{split}
det(A) & = ((1*2*1)+(-1*-\alpha*\alpha)+(\alpha*-1*-1))-((-1*-1* 1)+(1*-\alpha*1)+(\alpha * 2 * \alpha ))\\
& = (2 + (\alpha)^2 - \alpha) - (2(\alpha)^2 - \alpha +1)\\
& = -(\alpha)^2 + 0 + 1\\
\end{split}
\end{equation*}
Si $det(A) = 0\\$
\begin{equation*}
\begin{split}
-(\alpha)^2 + 1 = 0\\
(\alpha)^2 = 1\\
\alpha  = \pm 1 
\end{split}
\end{equation*}

Luego, sustituyendo los valores de 1 y -1 en la matriz ampliada $A|b$ y realizando operaciones de fila tenemos que:
- Si $a = 1$, entonces el sistema no tiene niguna solución.
- Si $a = -1$, entonces el sistema tiene infinitas soluciones. 

**c. Suponga que existe una única solución para un $\alpha$ determinado, encuentre dicha solución.**

De los literales anteriores sabemos que, si $\alpha \ \pm \neq 1$, entonces el sistema tiene solución única.
Ahora, tomando $\alpha = 0$ y aplicando operaciones de fila en la matriz ampliada $A|b$ llegamos a:
$\\x_1 = -1, x_2 = 1, x_3 = 1$



```python
matriz_ampliada = [[1, -1, 0, -2],[-1, 2, 0, 3],[0, 1, 1, 2]]
eliminacion_gaussiana(matriz_ampliada)
```
