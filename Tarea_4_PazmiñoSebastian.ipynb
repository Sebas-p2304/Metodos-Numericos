*1. Use el método de la bisección para encontrar soluciones precisas dentro de $10^-2$ para $x^3 -7x^2 +14x - 6 = 0$ en cada intervalo*
- $ a.[0,1]$
- $ b.[1,3.2]$
- $ c.[3.2,4]$

```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6  

def biseccion(a, b, tol=1e-2, max_iter=50):
    print(f"\nIntervalo inicial: [{a}, {b}]")
    if f(a) * f(b) >= 0:
        print("El método de biseccion no se puede aplicar:" )#f(a) y f(b) deben tener signos opuestos.
        return None

    iteracion = 0
    while abs(b - a) > tol and iteracion < max_iter:
        c = (a + b) / 2.0

        if f(c) == 0:
            return c  # algoritmo para encontrar raiz exacta 
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1




        print(f"Iteracion {iteracion}: c = {c}, f(c) = {f(c)} rango = [{a}, {b}]")
    return (a + b) / 2.0

# 1er intervalo de prueba
raiz = biseccion(0, 1)
print(f"\nRaiz aproximada: {raiz}")

# 2do intervalo de prueba
raiz = biseccion(1, 3.2)
print(f"\nRaiz aproximada: {raiz}")

# 3er intervalo de prueba 
raiz = biseccion(3.2, 4)
print(f"\nRaiz aproximada: {raiz}")
```

*2.  Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa 𝑚 cae desde una altura  $S_0$  y que la altura del objeto después de 𝑡 segundos es*

*$ S(t) = S_0 - \frac{mg}{k} t + \frac{m^{2}g}{k^{2}} ( 1 - e^{-kt/m} ) $*

*donde g = 9.81 $m/s^{2}$ y k representa el coeficiente de la resistencia del aire en $Ns/m$. Suponga $S_0$ = 300m, m = 0.25 kg y k = 0.1 Ns/m. Encuentre dentro de 0.01 𝑠𝑒𝑔𝑢𝑛𝑑𝑜𝑠, el tiempo que tarda un cuarto de kg en golpear el piso.*



Datos: $ g = 9.81 m/s^{2}  , S_0 = 300m ,  m = 0.25 kg , k = 0.1 Ns/m, dentro de 0.01 𝑠𝑒𝑔𝑢𝑛𝑑𝑜𝑠  $


Ecuación del volumen: 

$ S(t) = S_0 - \frac{mg}{k} t + \frac{m^{2}g}{k^{2}} ( 1 - e^{-kt/m} )  $

Sustituyendo los valores: 

$ 0 = 300 - \frac{(0.25)(9.81)}{0.1} t + \frac{(0.25)^{2}(9.81)}{0.1^{2}} ( 1 - e^{-0.1t/0.25} )  $

simplificando

$0 = 300 - \frac{2.4525}{0.1} t + \frac{0.613125}{0.01} ( 1 - e^{-0.4t} )$

Ecuación final a resolver:
$ 0 = 300 - 24.525 t + 61.3125 ( 1 - e^{-0.4t} )  $

```python
def f(x):
    return 300 -24.525 * x + 61.3125 * ( 1 - np.e**(-0.4*x) )

def biseccion(a, b, tol=1e-2, max_iter=50):
    print(f"\nIntervalo inicial: [{a}, {b}]")
    if f(a) * f(b) >= 0:
        print("El método de bisección no se puede aplicar: f(a) y f(b) deben tener signos opuestos.")
        return None

    iteracion = 0
    while abs(b - a) >= tol and iteracion < max_iter:
        c = (a + b) / 2.0

        if f(c) == 0:
            return c  # Se encontró la raíz exacta
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1

        print(f"Iteración {iteracion}: c = {c}, f(c) = {f(c)} rango = [{a}, {b}]","Error:",abs(b - a) )
    return (a + b) / 2.0

# Intervalos
raiz = biseccion(14,15,)
print(f"\nRaíz aproximada: t = {raiz}")
```
