# Mínimos cuadrados
## Prueba 02
Interpole los siguientes conjuntos de datos con la función correspondiente. 

La ecuación de la línea es:

$ y(x) = a_1 x + a_0 $

Al realizar el proceso de mínimos cuadrados queda el siguiente sistema de ecuaciones:


$( \sum_i (y_i - a_1 x_i - a_0) , \sum_i (y_i - a_1 x_i - a_0)x_i ) = 0$

```python
# Derivadas parciales para regresión cuadrática
# y = a x^2 + b x + c

def der_parcial_a(xs: list, ys: list) -> tuple[float, float, float, float]:

    c_a = sum(x**4 for x in xs)
    c_b = sum(x**3 for x in xs)
    c_c = sum(x**2 for x in xs)
    c_ind = sum((x**2) * y for x, y in zip(xs, ys))

    return (c_a, c_b, c_c, c_ind)


def der_parcial_b(xs: list, ys: list) -> tuple[float, float, float, float]:

    c_a = sum(x**3 for x in xs)
    c_b = sum(x**2 for x in xs)
    c_c = sum(x for x in xs)
    c_ind = sum(x * y for x, y in zip(xs, ys))

    return (c_a, c_b, c_c, c_ind)


def der_parcial_c(xs: list, ys: list) -> tuple[float, float, float, float]:
 
    c_a = sum(x**2 for x in xs)
    c_b = sum(x for x in xs)
    c_c = len(xs)
    c_ind = sum(ys)

    return (c_a, c_b, c_c, c_ind)
```

```python
import math

def der_parcial_exp_a(xs: list, ys: list, a: float, b: float) -> float:
   
    return sum(
        (a * math.exp(b * x) - y) * math.exp(b * x)
        for x, y in zip(xs, ys)
    )


def der_parcial_exp_b(xs: list, ys: list, a: float, b: float) -> float:
 
    return sum(
        (a * math.exp(b * x) - y) * a * x * math.exp(b * x)
        for x, y in zip(xs, ys)
    )
```
