### Ejercicio 1 

### Grafique la trayectorio de los siguientes sistemas de ecuaciones 

$X_1 +X_2 = 7$

$-2X_1 +5X_2 = 0 $

$X_0 = (0,0)$

$X_0 = (5,2)$

```python
A = np.array([[1,1], [-2,5]])
b = np.array([7,0])
n = 2
gauss_seidel(A=A, b=b, x=[0] * n, tol=1e-5, max_iter=10, return_traj=True)  # type: ignore
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sistema
A = np.array([[1, 1],
              [-2, 5]], dtype=float)

b = np.array([7, 0], dtype=float)
n = 2

# Punto inicial
x0 = np.array([0, 0], dtype=float)

# Ejecutar la trayectoria de Gauss-Seidel 
sol, traj = gauss_seidel(
    A=A,
    b=b,
    x=x0,
    tol=1e-5,
    max_iter=20,
    return_traj=True
)

# puntos de la trayectoria
px = []
py = []

for k in range(len(traj) - 1):
    x_prev = traj[k]
    x_next = traj[k + 1]

    # Paso 1: actualizar x1 (horizontal)
    px.append(x_prev[0])
    py.append(x_prev[1])

    px.append(x_next[0])
    py.append(x_prev[1])

    # Paso 2: actualizar x2 (vertical)
    px.append(x_next[0])
    py.append(x_next[1])

# grafica de la trayectoria
plt.figure()
plt.plot(px, py, '-o', color='green')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("trayectoria con el Metodo de Gauss-Seidel")
plt.grid(True)
plt.axis("equal")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sistema
A = np.array([[1, 1],
              [-2, 1]], dtype=float)

b = np.array([6, 0], dtype=float)
n = 2

# Punto inicial
x0 = np.array([-3, 2], dtype=float)

# Ejecutar Gauss-Seidel con trayectoria
sol, traj = gauss_seidel(
    A=A,
    b=b,
    x=x0,
    tol=1e-5,
    max_iter=10,
    return_traj=True
)

# puntos de la trayectoria
px = []
py = []

for k in range(len(traj) - 1):
    x_prev = traj[k]
    x_next = traj[k + 1]

    # Paso 1: actualizar x1 (horizontal)
    px.append(x_prev[0])
    py.append(x_prev[1])

    px.append(x_next[0])
    py.append(x_prev[1])

    # Paso 2: actualizar x2 (vertical)
    px.append(x_next[0])
    py.append(x_next[1])

# grafica de la trayectoria 
plt.figure()
plt.plot(px, py, '-o', color='green')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Trayectoria con el  Metodo de Gauss-Seidel")
plt.grid(True)
plt.axis("equal")
plt.show()
```

Pruebe 3 posiciones iniciales, encuentre una en la que diverge el sistema 

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz del sistema
A = np.array([
    [ 1, -1,  2,  0],
    [-1,  1, -1,  3],
    [ 2, -1,  1, -1],
    [ 0,  3, -1,  2]
], dtype=float)

b = np.array([1360, 1130, 1350, 3650], dtype=float)

n = 4
x0 = [0] * n   # aproximación inicial

# Ejecutar Gauss-Seidel con trayectoria
sol, traj = gauss_seidel(
    A=A,
    b=b,
    x=x0,
    tol=1e-5,
    max_iter=25,
    return_traj=True
)

print("Solución aproximada:")
print(sol.ravel())

# Número de iteraciones
iters = np.arange(traj.shape[0])

# Gráfica de la trayectoria (componentes vs iteración)
plt.figure()
plt.plot(iters, traj[:, 0], marker="o", label="x1")
plt.plot(iters, traj[:, 1], marker="o", label="x2")
plt.plot(iters, traj[:, 2], marker="o", label="x3")
plt.plot(iters, traj[:, 3], marker="o", label="x4")

plt.xlabel("Iteracion k")
plt.ylabel("Valor de la variable")
plt.title("Trayectoria con el metodo de gauss-seidel ")
plt.grid(True)
plt.legend()
plt.show()
```
