*2. Calcule los errores absoluto y relativo en las aproximaciones de  $p$ por $p^*$*

- a. $p = e^{10} , p^* = 22000$

- b. $p = 10^{\pi} , p^* = 1400$

- c. $p = 8! , p^* = 39900$

- d. $p! , p^s* = \sqrt{18\pi}(9/e)^{9} $

 

```python
import math

constante_euler = math.e
pi = 3.14159265

# ********** a **********   
valor_real_a = constante_euler ** 10
valor_aproximado_a = 22000

error_absoluto_a = abs(valor_real_a - valor_aproximado_a)
error_relativo_a = abs((valor_real_a - valor_aproximado_a) / valor_real_a)
print("error absoluto =", error_absoluto_a, "error relativo =", error_relativo_a)

# ********* b **********     
valor_real_b = 10 ** pi
valor_aproximado_b = 1400

error_absoluto_b = abs(valor_real_b - valor_aproximado_b)
error_relativo_b = abs((valor_real_b - valor_aproximado_b) / valor_real_b)
print("error absoluto =", error_absoluto_b, "error relativo =", error_relativo_b)

# ******** c **********     
valor_real_c = math.factorial(8)
valor_aproximado_c = 39900

error_absoluto_c = abs(valor_real_c - valor_aproximado_c)
error_relativo_c = abs((valor_real_c - valor_aproximado_c) / valor_real_c)
print("error absoluto =", error_absoluto_c, "error relativo =", error_relativo_c)

# ******** d **********     
valor_real_d = math.factorial(9)
valor_aproximado_d = math.sqrt(18 * pi) * (9 / constante_euler) ** 9

error_absoluto_d = abs(valor_real_d - valor_aproximado_d)
error_relativo_d = abs((valor_real_d - valor_aproximado_d) / valor_real_d)
print("error absoluto =", error_absoluto_d, "error relativo =", error_relativo_d)
```

*5. Los primeros tres t√©rminos diferentes a cero de la serie de Maclaurin para la funci√≥n arcotangente son: $ x - (1/3)x^{3} + (1/5)x^{5} $ Calcule los errores absoluto y relativo en las siguientes aproximaciones de ùúã mediante el polinomio en lugar del arcotangente:*

**-a. $ 4 [ arctan (\frac{1}{2}) + arctan (\frac{1}{3})]$**

**-b. $ 16  arctan (\frac{1}{5}) - 4 arctan (\frac{1}{239})$**

```python
import math

pi = math.pi

# *************** a ****************
print("a)")
valor_x1_a = 1/2
termino_arctan_x1_a = valor_x1_a - (1/3)*valor_x1_a**3 + (1/5)*valor_x1_a**5

valor_x2_a = 1/3
termino_arctan_x2_a = valor_x2_a - (1/3)*valor_x2_a**3 + (1/5)*valor_x2_a**5

valor_real_pi = pi
aproximacion_pi_a = 4 * (termino_arctan_x1_a + termino_arctan_x2_a)

print("valor real:", valor_real_pi, "valor aproximado:", aproximacion_pi_a)

error_absoluto_a = abs(valor_real_pi - aproximacion_pi_a)
error_relativo_a = abs((valor_real_pi - aproximacion_pi_a) / valor_real_pi)
print("error absoluto =", error_absoluto_a, "error relativo =", error_relativo_a)

# *************** b ****************
print("b)")
valor_x1_b = 1/5
termino_arctan_x1_b = valor_x1_b - (1/3)*valor_x1_b**3 + (1/5)*valor_x1_b**5

valor_x2_b = 1/239
termino_arctan_x2_b = valor_x2_b - (1/3)*valor_x2_b**3 + (1/5)*valor_x2_b**5

valor_real_pi = pi
aproximacion_pi_b = (16 * termino_arctan_x1_b - 4 * termino_arctan_x2_b)

print("valor real:", valor_real_pi, "valor aproximado:", aproximacion_pi_b)

error_absoluto_b = abs(valor_real_pi - aproximacion_pi_b)
error_relativo_b = abs((valor_real_pi - aproximacion_pi_b) / valor_real_pi)
print("error absoluto =", error_absoluto_b, "error relativo =", error_relativo_b)
```

*7. Suponga que dos puntos $(x_0 , y_0) (x_1 , y_1)$  se encuentran en l√≠nea recta con $y_1 \ne y_0$. Existen dos f√≥rmulas para encontrar la intersecci√≥n $ùë•$ de la l√≠nea:*

$ x = \frac{x_0y_1 - x_1y_0}{y_1 - y_0} $ y $ x = x_0 - \frac{(x_1 - x_0)y_0}{y_1 - y_0}$

*a. Use los datos $(x_0 , y_0) = (1.31 , 3.24)$ y $(x_1 , y_1) = (1.93 , 5.76)$  y la aritm√©tica de redondeo de tres d√≠gitos para calcular la intersecci√≥n con $x$ de ambas maneras. ¬øCu√°l m√©todo es mejor y por qu√©?*

### *Formula 1.* 

$ x = \frac{x_0y_1 - x_1y_0}{y_1 - y_0} $

$ x = \frac{(1.31) \cdot (5.76) - (1.93)\cdot (3.24)}{5.76 - 3.24} $

$ x = \frac{7.546 - 6.253}{2.52} $

$ x = \frac{1.293}{2.52} = 0.513$ 

### *Formula 2.*

$ x = x_0 - \frac{(x_1 - x_0)y_0}{y_1 - y_0}$

$ x = 1.31 - \frac{(1.93 - 1.31)3.24}{5.76 - 3.24}$

$ x = 1.31 - \frac{(0.62)3.24}{2.52}$
    
$ x = 1.31 - \frac{2.009}{2.52}$

$ x = 1.31 - 0.797 = 0.513$

```python
x0= 1.31
y0= 3.24

x1= 1.93 
y1= 5.76

# primera f√≥rmula
a = (x0*y1 - x1*y0)/(y1-y0)
# Segunda f√≥rmula
b=  x0 - ((x1-x0)*y0)/(y1-y0)


print("Primera Formula", round(a,3))
print("Segunda Formula", round(b,3))
```

*La segunda f√≥rmula opera de manera m√°s eficiente y con mayor estabilidad num√©rica. Tambi√©n se observa que reduce la propagaci√≥n de errores, como aquellos derivados del redondeo durante los c√°lculos. Esta aproximaci√≥n evita operaciones potencialmente problem√°ticas como ciertas multiplicaciones que pueden amplificar los errores. Adem√°s, se destaca por su estructuraci√≥n de operaciones dise√±ada para preservar la precisi√≥n en los c√°lculos.*

*Es importante se√±alar que ambas f√≥rmulas son equivalentes en teor√≠a y producen el mismo resultado exacto. Sin embargo, podr√≠an generar diferencias num√©ricas cuando se trabajan con un mayor n√∫mero de d√≠gitos o en implementaciones computacionales espec√≠ficas, debido a la distinta forma en que manejan los errores de precisi√≥n.*
